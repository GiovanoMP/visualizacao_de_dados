**Visualização de Dados**

Vamos criar visualizações utilizando apenas a biblioteca Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'meses' :['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    'produto1_vendas' : np.random.randint(10, 50, 12),
    'produto2_vendas' : np.random.randint(20, 60, 12),
    'produto3_vendas' : np.random.randint(5, 30, 12)
}

#transformando o dicioário em um Data Frame
df_vendas = pd.DataFrame(data)

#Salvando o arquivo no diretório em csv
df_vendas.to_csv("dados_vendas.csv", index=False)
```

```python
dados_vendas = pd.read_csv("dados_vendas.csv")
dados_vendas.head
```

**Vamos criar gráficos de linha para mostrar as vendas mensais de diferentes produtos. Usaremos cores, marcadores e tracejados para distinguir os produtos, e acrescentaremos rótulos, títulos e legendas para tornar os gráficos informativos.**

```python

# Gráfico de linha combinado com o DF dados_vendas

#Definindo tamanho do gráfico
plt.figure(figsize=(10,6))
plt.plot(dados_vendas['meses'], dados_vendas['produto1_vendas'], color='blue', marker='o', linestyle='-', label='Produto 1')
plt.plot(dados_vendas['meses'], dados_vendas['produto2_vendas'], color='red', marker='s', linestyle='--', label='Produto 2')
plt.plot(dados_vendas['meses'], dados_vendas['produto3_vendas'], color='green', marker='^', linestyle='-.', label='Produto 3')

#Definindo título e labels
plt.title("Vendas Mensais dos Produtos")
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.legend()

#Utilizando grid e tight_layout
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

**Vamos criar subplots para mostrar as vendas de diferentes produtos lado a lado, experimentando diferentes estilos e layouts de gráficos utilizando a biblioteca Matplotlib.**

```python
# Criação de 3 gráficos lado a lado com estilos diferentes
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18,6))

# Gráfico Produto 1
axes[0].bar(dados_vendas['meses'], dados_vendas['produto1_vendas'], color='blue', alpha=0.7)
axes[0].set_title('Vendas Produto 1')
axes[0].set_xlabel('Meses')
axes[0].set_ylabel('Vendas')
axes[0].grid(axis='y', linestyle='--', linewidth=0.5)

# Gráfico Produto 2
axes[1].scatter(dados_vendas['meses'], dados_vendas['produto2_vendas'], color='red', s=100, edgecolor='black')
axes[1].set_title('Vendas Produto 2')
axes[1].set_xlabel('Meses')
axes[1].set_ylabel('Vendas')
axes[1].grid(axis='y', linestyle='--', linewidth=0.5)

# Gráfico Produto 3
axes[2].plot(dados_vendas['meses'], dados_vendas['produto3_vendas'], color='green', marker='^', linestyle='-.')
axes[2].fill_between(dados_vendas['meses'], dados_vendas['produto3_vendas'], color='green', alpha=0.2)
axes[2].set_title('Vendas Produto 3')
axes[2].set_xlabel('Meses')
axes[2].set_ylabel('Vendas')
axes[2].grid(axis='y', linestyle='--', linewidth=0.5)

# Ajustar layout
plt.tight_layout()
plt.show()
```

* Gráficos de barras são usados para comparar valores entre categorias. Eles são úteis para mostrar a distribuição de dados ou para comparar dados entre grupos diferentes.

* Gráficos de linhas são usados para mostrar tendências ao longo do tempo. Eles são úteis para mostrar como os dados mudam com o tempo ou para comparar dados entre diferentes períodos de tempo.

Geralmente, devemos escolher um gráfico de barras quando estivermos comparando valores entre categorias e um gráfico de linhas quando estiverermos mostrando tendências ao longo do tempo.

**Vamos criar visualizações utilizando as bibliotecas Pandas, Seaborn e Matplotlib com os dados de consumo de energia elétrica.**

Dados de Consumo de Energia Elétrica

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
consumo = np.random.uniform(100, 500, len(data))

df_consumo = pd.DataFrame({'Data': data, 'Consumo': consumo})

#Salvando o arquivo em csv
df_consumo.to_csv('dados_consumo.csv', index=False)
```

- Pandas: É uma biblioteca de análise de dados que fornece uma estrutura de dados chamada DataFrame. Os DataFrames são uma forma conveniente de armazenar e manipular dados tabulares, como planilhas. O Pandas também fornece uma variedade de funções para visualizar dados, incluindo gráficos de barras, gráficos de linhas, gráficos de dispersão e muito mais.

- Seaborn: É uma biblioteca de visualização de dados que é construída sobre o Matplotlib. O Seaborn fornece uma interface de alto nível para criar gráficos bonitos e informativos. O Seaborn também fornece uma variedade de funções para estilizar e personalizar gráficos.

- Matplotlib: É uma biblioteca de visualização de dados de baixo nível que fornece uma ampla gama de recursos para criar gráficos. O Matplotlib é flexível e poderoso, mas pode ser difícil de aprender. Além disso, seus gráficos são considerados mais simples e com menos recursos se comparados ao Seaborn por exemplo.

**Vamos usar Pandas para criar gráficos de linha, barras e histogramas para mostrar o consumo de energia ao longo do tempo.**

```python
dados_consumo = pd.read_csv("dados_consumo.csv")
dados_consumo.head()
```

```python
# Gráfico de Linha para mostrar o consumo ao longo do tempo
plt.figure(figsize=(12,6))
dados_consumo.plot(x='Data', y='Consumo', kind='line', ax=plt.gca(), color='blue', title="Consumo de Energia ao Longo do Tempo")
plt.ylabel("Consumo")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de Barras para mostrar o consumo mensal médio

# Convertendo a coluna 'Data' para datetime
dados_consumo['Data'] = pd.to_datetime(dados_consumo['Data'])

dados_consumo['Mês'] = dados_consumo['Data'].dt.month
consumo_mensal_medio = dados_consumo.groupby('Mês')['Consumo'].mean()
plt.figure(figsize=(12,6))
consumo_mensal_medio.plot(kind='bar', ax=plt.gca(), color='green', title="Consumo Médio Mensal de Energia")
plt.xlabel("Mês")
plt.ylabel("Consumo Médio")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Histograma para mostrar a distribuição do consumo
plt.figure(figsize=(12,6))
dados_consumo['Consumo'].plot(kind='hist', bins=30, ax=plt.gca(), color='red', edgecolor='black', title="Distribuição do Consumo de Energia")
plt.xlabel("Consumo")
plt.ylabel("Frequência")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

**Vamos utilizar Seaborn para criar gráficos de dispersão, gráficos hexagonais e gráficos de pizza com informações de consumo de energia elétrica.**

```python
# Configuração de estilo
sns.set_style("whitegrid")

# Gráfico de Dispersão
plt.figure(figsize=(12,6))
sns.scatterplot(data=dados_consumo, x='Data', y='Consumo', color='blue', alpha=0.6)
plt.title("Gráfico de Dispersão do Consumo de Energia")
plt.show()
```

```python
# Preparando dados para o Gráfico de Pizza
# Categorizando o consumo em 'Baixo', 'Médio' e 'Alto'
bins = [100, 300, 400, 500]
labels = ['Baixo', 'Médio', 'Alto']
dados_consumo['Categoria'] = pd.cut(dados_consumo['Consumo'], bins=bins, labels=labels, right=False)
categoria_counts = dados_consumo['Categoria'].value_counts()

# Gráfico de Pizza
plt.figure(figsize=(8,8))
categoria_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['green', 'yellow', 'red'], explode=(0.1, 0, 0))
plt.title("Distribuição do Consumo de Energia por Categoria")
plt.ylabel('')  # Remove o label do eixo y
plt.show()
```

```python
%%writefile dados_esportes.py

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from streamlit_folium import folium_static

# Carregar o DataFrame dados_esportes do arquivo CSV
dados_esportes = pd.read_csv('dados_esportes.csv')

st.title('Estatísticas de Esportes')

# Primeiros Registros
st.write(dados_esportes.head())

# Gráfico de barras com a média de pontos por esporte
st.subheader('Média de pontos por esporte:')
fig, ax = plt.subplots()
sns.barplot(data=dados_esportes, x='Esporte', y='Pontos', ax=ax)
st.pyplot(fig)

# Histograma de jogos disputados usando Matplotlib diretamente
st.subheader('Distribuição de jogos disputados:')
fig, ax = plt.subplots()
ax.hist(dados_esportes['Jogos Disputados'], bins=20, edgecolor='black', alpha=0.7)
ax.set_title('Distribuição de Jogos Disputados')
ax.set_xlabel('Jogos Disputados')
ax.set_ylabel('Número de Times')
st.pyplot(fig)

# Criando um mapa usando Folium
m = folium.Map(location=[20, 0], zoom_start=2)
for city in dados_esportes['Cidade'].unique():
    folium.Marker(location=[0, 0], popup=city).add_to(m) 

st.subheader('Localização das equipes esportivas usando Folium:')
folium_static(m)

# Seletor de esporte
esporte_selecionado = st.selectbox('Selecione um esporte para visualizar estatísticas:', dados_esportes['Esporte'].unique())

# Exibindo estatísticas do esporte selecionado
dados_filtrados = dados_esportes[dados_esportes['Esporte'] == esporte_selecionado]
st.write(dados_filtrados.describe())
```

```python
#Conjunto de Dados: Dados de Pesquisas de Satisfação do Cliente

import pandas as pd

clientes = ['Cliente 1', 'Cliente 2', 'Cliente 3', 'Cliente 4', 'Cliente 5']

idade = [25, 30, 22, 40, 35]

satisfacao = [4, 5, 3, 4, 2]

dados_satisfacao = pd.DataFrame({'Cliente': clientes, 'Idade': idade, 'Satisfação': satisfacao})
dados_satisfacao.to_csv('dados_satisfacao.csv', index=False)
```

**Aplicativo Streamlit:**

Vamos configurar uma interface para o upload de arquivos de pesquisas de satisfação usando o Streamlit. Organizaremos os elementos em colunas para melhorar o layout do aplicativo. Além disso, vamos explorar a criação de aplicativos com várias páginas e a navegação entre elas. Acrescentaremos interatividade para realizar análises de dados com base nas respostas da pesquisa.

```python
%%writefile satisfacao.py

import streamlit as st
import pandas as pd

#Criando menu lateral com 3 opções
def main():
    menu = ["Upload", "Análise", "Sobre"]
    choice = st.sidebar.selectbox("Escolha uma opção", menu)

#Opção de escolha do usuário para upload
    if choice == "Upload":
        st.subheader("Upload de Arquivos de Pesquisas de Satisfação")
        
        file = st.file_uploader("Escolha um arquivo CSV", type=["csv"])
        if file is not None:
            df = pd.read_csv(file)
            st.write(df)
    #Opção de Análise        
    elif choice == "Análise":
        st.subheader("Análise de Dados de Satisfação")
        
        file = st.file_uploader("Escolha um arquivo CSV", type=["csv"])
        if file is not None:
            df = pd.read_csv(file)
            
            # Usando st.columns para melhorar o layout
            col1, col2 = st.columns(2)
            
            with col1:
                if st.checkbox("Mostrar dados"):
                    st.write(df)
            
            with col2:
                if st.checkbox("Mostrar média de satisfação"):
                    st.write(df["Satisfação"].mean())

            
#Opção sobre:
    elif choice == "Sobre":
        st.subheader("Sobre o Aplicativo")
        st.write("Este é um aplicativo simples para analisar pesquisas de satisfação.")

if __name__ == "__main__":
    main()
```
